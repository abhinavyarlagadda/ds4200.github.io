**V3:** Boston ozone vs. NO₂ concentration (2025)

```python
import pandas as pd
import altair as alt

alt.data_transformers.disable_max_rows()
```

CO Concentration Over Time by City (2025)

Scatter plot of daily CO readings for five major U.S. cities, colored by city.
Allows direct comparison of CO levels and seasonal trends across different metro areas.

```python
# --- Load CO data & filter to five cities ---
co = pd.read_csv("daily_42101_2025.csv", low_memory=False)
co.columns = co.columns.str.strip()

cities = ["Boston", "New York", "Washington", "Los Angeles", "Houston", "Atlanta"]

# Note: DC monitors are listed under 'Washington' in City Name
city_co = co[
    co["City Name"].str.contains("|".join(cities), case=False, na=False)
].copy()

# Normalize city labels
def label_city(name):
    name = str(name)
    if "Boston"      in name: return "Boston"
    if "New York"    in name: return "New York"
    if "Washington"  in name: return "Washington, DC"
    if "Los Angeles" in name: return "Los Angeles"
    if "Houston"     in name: return "Houston"
    if "Atlanta"     in name: return "Atlanta"
    return None

city_co["City"] = city_co["City Name"].apply(label_city)
city_co = city_co.dropna(subset=["City"])
city_co["Date Local"] = pd.to_datetime(city_co["Date Local"])

# Daily mean across all monitors per city
daily_city = (
    city_co.groupby(["Date Local", "City"], as_index=False)
    .agg(co_ppm=("Arithmetic Mean", "mean"), aqi=("AQI", "mean"))
    .sort_values("Date Local")
)

daily_city.head()
```

```python
# --- Build V3 chart ---

city_order = ["Boston", "New York", "Washington, DC", "Los Angeles", "Houston", "Atlanta"]

# Per-city LOESS trend lines
trend = alt.Chart(daily_city).transform_loess(
    "Date Local", "co_ppm", groupby=["City"], bandwidth=0.4
).mark_line(
    strokeWidth=1.5, opacity=0.5, strokeDash=[4, 2]
).encode(
    x="Date Local:T",
    y="co_ppm:Q",
    color="City:N"
)

# Scatter points colored by city
points = alt.Chart(daily_city).mark_circle(size=55, opacity=0.75).encode(
    x=alt.X(
        "Date Local:T",
        title="Date",
        axis=alt.Axis(format="%b", tickCount="month", labelAngle=0)
    ),
    y=alt.Y(
        "co_ppm:Q",
        title="Daily Avg CO (ppm)",
        scale=alt.Scale(zero=False)
    ),
    color=alt.Color(
        "City:N",
        sort=city_order,
        scale=alt.Scale(scheme="category10"),
        legend=alt.Legend(orient="right", title="City")
    ),
    tooltip=[
        alt.Tooltip("Date Local:T", title="Date",     format="%b %d, %Y"),
        alt.Tooltip("City:N",       title="City"),
        alt.Tooltip("co_ppm:Q",     title="CO (ppm)", format=".3f"),
        alt.Tooltip("aqi:Q",        title="Avg AQI",  format=".1f")
    ]
)

v3 = alt.layer(trend, points).properties(
    title=alt.TitleParams(
        text="Daily CO Concentration Over Time by City (2025)",
        subtitle="Daily avg CO across monitors in six major U.S. cities, colored by city",
        fontSize=16,
        subtitleFontSize=12,
        anchor="start"
    ),
    width=900,
    height=420
).interactive()

v3
```

```python
v3.save("co_by_city_over_time.html")
```
